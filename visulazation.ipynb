```python
from vasicek_model import Vasicek
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plot(data):
    if data.ndim == 2:
        plt.plot(np.arange(len(data)), np.mean(data, axis=1), color='orange')
    else:
        plt.plot(np.arange(len(data)), data, color='orange')
```
